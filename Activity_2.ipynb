```python
#Activity 2
import sqlite3
import pandas as pd

conn = sqlite3.connect("school.db")
```

```python
students = pd.read_csv("students.csv")
courses  = pd.read_csv("courses.csv")
enrollments = pd.read_csv("enrollments.csv")
```

```python
students.to_sql("students", conn, if_exists="replace", index=False)
courses.to_sql("courses", conn, if_exists="replace", index=False)
enrollments.to_sql("enrollments", conn, if_exists="replace", index=False)
```

```python
#Task1
pd.read_sql_query("SELECT * FROM students", conn)
pd.read_sql_query("SELECT * FROM courses", conn)
pd.read_sql_query("SELECT * FROM enrollments", conn)
```

```python
#Task2
pd.read_sql_query("SELECT * FROM students WHERE city = 'Cairo'",conn)
```

```python
#Task3
pd.read_sql_query("SELECT name, city, score FROM students WHERE score >= 90 ORDER BY score DESC",conn)
```

```python
#Task4
pd.read_sql_query("SELECT name, score FROM students WHERE city = 'Giza' ORDER BY score DESC LIMIT 3",conn)
```

```python
#Task5
pd.read_sql_query("SELECT name, score FROM students ORDER BY score DESC LIMIT 4 OFFSET 3",conn)
```
