```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("school.db")
```

```python
students = pd.read_csv("students.csv")
students
```

```python
students.to_sql("students", conn, if_exists="replace", index=False)
```

```python
pd.read_sql_query("SELECT * FROM students", conn)
```

```python
pd.read_sql_query("SELECT name, city, score FROM students", conn)
```

```python
pd.read_sql_query("SELECT name, city, score FROM students WHERE city = 'Cairo'", conn)
```

```python
pd.read_sql_query("SELECT name  , score from students WHERE score >= 85",conn)
```

```python
pd.read_sql_query("SELECT name , city , score from students WHERE city = 'Giza' AND score >= 80",conn)
```

```python
pd.read_sql_query("SELECT name, score FROM students ORDER BY score DESC", conn)
```

```python
pd.read_sql_query("SELECT name, score FROM students ORDER BY score DESC LIMIT 3", conn)
```

```python
pd.read_sql_query("SELECT name, score FROM students ORDER BY score DESC LIMIT 4 OFFSET 4", conn)
```

```python
pd.read_sql_query("SELECT name,city , score FROM students WHERE city = 'Cairo' ORDER BY score DESC LIMIT 3", conn)
```
